Plotting the graph for different $t_{thresholds}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants
alpha = 1/137
kn = -1.191
mn = 939.6
t_min = 1e-6
t_max = 1e-2
```

```python
# Differential cross section

def dsig_dt(t):
    return ((alpha * kn)/mn)**2 * (np.pi/abs(t))


# Sigma
def sigma(t):
    return ((alpha * kn)/mn)**2 * np.pi * np.log(t/t_min)
```

```python
t_values = np.logspace(np.log10(t_min), np.log10(t_max), 500)
dsig_dt_vals = dsig_dt(t_values)
sig_vals = sigma(t_values)
```

```python
plt.plot(t_values, dsig_dt_vals)
plt.title("Differential Cross Section")
plt.xlabel("$t$ (GeV)")
plt.ylabel("$d\sigma/dt$ (mb/GeV$^{-1}$)")
```

```python
plt.plot(t_values, sig_vals)
plt.title("Total Cross Section")
plt.xlabel("$t$ (GeV)")
plt.ylabel("$\sigma$ (mb)")
```
